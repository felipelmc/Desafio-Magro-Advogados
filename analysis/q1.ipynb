```python
import pandas as pd
from pyvis.network import Network
```

```python
conexoes = pd.read_excel("../data/network.xlsx", sheet_name="Conexões")
fotos = pd.read_excel("../data/network.xlsx", sheet_name="Fotos")
cargos = pd.read_excel("../data/network.xlsx", sheet_name="Cargos")
```

```python
conexoes
```

```python
fotos
```

```python
cargos
```

```python
net = Network(notebook=False,
              cdn_resources="remote",
              bgcolor="white", 
              font_color="black", 
              height="750px", 
              width="100%",
              select_menu=True)

nodes = cargos["stakeholder"].tolist()
edges = conexoes[["stakeholder", "connection"]].values.tolist()

for stakeholder in nodes:
    net.add_node(stakeholder, 
                 title=cargos[cargos["stakeholder"] == stakeholder]["cargo"].values[0], 
                 shape="image", 
                 image=fotos[fotos["stakeholder"] == stakeholder]["foto"].values[0],
                 size=50,
                 physics=True)
    
net.add_edges(edges)

net.repulsion(node_distance=300,
              spring_length=100,
              spring_strength=0.01)

# save to html file
net.save_graph("../networks/net-geral.html")
```
